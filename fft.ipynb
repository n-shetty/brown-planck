```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = 1; D = 1e1 # arbitrary values
t = np.linspace(start=1e-3, stop=10, num=100000) # time
y = x/t * 1/2 * 1/np.sqrt(3.14*D*t) * np.exp(-1/(4*D*t)*np.square(x)) # Equation number 32 from the paper: 
                                                                      # https://facultystaff.richmond.edu/~olipan/Chandrashekhar.pdf
                                                                      # corresponding to random walk with absorbing wall @ x
```

```python
plt.plot(t,y)
```

```python
plt.loglog(1/t,y) # plot freq for fun
```

```python
Y = np.fft.fft(y)
```

```python
N = int(len(Y)/2+1)
Y[N-4:N+3]
```

```python
dt = t[1] - t[0]
fa = 1.0/dt # scan frequency
print('dt=%.5fs (Sample Time)' % dt)
print('fa=%.2fHz (Frequency)' % fa)
```

```python
X = np.linspace(0, fa/2, N, endpoint=True)
```

```python
plt.plot(X, 2.0*np.abs(Y[:N])/N)
plt.xlabel('Frequency ($Hz$)')
plt.ylabel('Amplitude ($Unit$)')
```

```python
plt.loglog(X, 2.0*np.abs(Y[:N])/N)
plt.xlabel('Frequency ($Hz$)')
plt.ylabel('Amplitude ($Unit$)')
```
